```python
# imports
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl import load_workbook
from vega_datasets import data as vds
import pandas as pd
import numpy as np
```

```python
# dataframe
cars = vds.cars()
cars.head()
```

```python
# dataframe to spreadsheet

# create a workbook and grab active worksheet
workbook = Workbook()
worksheet = workbook.active

# loop over dataframe values and append to worksheet
for i in dataframe_to_rows(cars, index=False, header=True):
    worksheet.append(i)
    
# save workbook
workbook.save('cars.xlsx')
```

```python
# spreadsheet to dataframe (with headers)

# load workbook
df_workbook = load_workbook('LibreOfficeCalcDoc.xlsx')

# grab active worksheet
df_worksheet = df_workbook.active

# create dataframe from worksheet values
df = pd.DataFrame(df_worksheet.values)

# grab the first row for the header
header = df.iloc[0]

# data without the header row
df = df.iloc[1:]

# set the header row as the df header
df.columns = header

df
```

```python
# list and array from column values

# create empty list
l = []

# loop over column values and append to list
for i in df_worksheet['A']:
    l.append(i.value)
    
# get rid of column header
l = l[1:]

# display list values
print(l)

# create numpy array from values
arr = np.array(l)

# use numpy array to get standard deviation
print(arr.std())
```
